```python
def generate_permutations(s):
    all_permutation = []

    if len(s) == 1:
        return [s]

    for i in range(len(s)):
        char = s[i]
        remain = s[:i] + s[i+1:]

        for j in generate_permutations(remain):
            new_per = char + j
            if new_per not in all_permutation:
                all_permutation.append(new_per)

    return all_permutation

print(generate_permutations("abc"))
```

```python
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}
def calculate_directory_size(directory):
    total_size = 0

    for key, value in directory.items():
        # If it's a file (size in KB)
        if isinstance(value, int):
            total_size += value
        # If it's a subdirectory
        elif isinstance(value, dict):
            total_size += calculate_directory_size(value)

    return total_size

# Test the function
total = calculate_directory_size(directory_structure)
print("Total directory size:", total, "KB")
```

```python
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0


    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1


coins = [1, 2, 5]
amount = 11
print(min_coins(coins, amount))
```

```python
def longest_common_subsequence(s1, s2):
    n = len(s1)
    m = len(s2)

    dp = [[0] * (m + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[n][m]
print(longest_common_subsequence("abcde", "ace"))

```

```python
def knapsack(weights, values, capacity):
    n = len(weights)

    # DP table: (n+1) x (capacity+1)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Build the table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                include = values[i - 1] + dp[i - 1][w - weights[i - 1]]
                exclude = dp[i - 1][w]
                dp[i][w] = max(include, exclude)
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

print(knapsack(weights, values, capacity))

```
