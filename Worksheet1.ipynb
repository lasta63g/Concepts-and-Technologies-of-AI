```python
import numpy as np

# Given arrays
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

print("1.(x + y):")
print(x + y)

print("\n2.(x - y):")
print(x - y)

print("\n3.(3 * x):")
print(3 * x)

print("\n4.(x**2):")
print(x**2)

print("\n5(a). Dot product of v.w:")
print(np.dot(v, w))

print("\n5(b). Dot product of x.v:")
print(np.dot(x, v))

print("\n5(c). Dot product of x .y:")
print(np.dot(x, y))

print("\n6(a). Concatenate x and y along rows:")
print(np.vstack((x, y)))

print("\n6(b). Concatenate v and w along columns:")
print(np.column_stack((v, w)))

print("\n7. Concatenate x and v (Shows error explanation):")
try:
    print(np.concatenate((x, v)))
except Exception as e:
    print("Error:", e)
    print("Reason: x is 2D but v is 1D. NumPy cannot concatenate arrays with different dimensions.")
```

```python
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# 1.
A_inv = np.linalg.inv(A)
I = A.dot(A_inv)

print("\n1. A * A^{-1} = I:",I)
print("Is Identity (approx)?", np.allclose(I, np.eye(2)))


# 2. Prove AB != BA
AB = A.dot(B)
BA = B.dot(A)

print("\n2. AB:\n", AB)
print("\n   BA:\n", BA)
print("\nIs AB = BA?", np.array_equal(AB, BA))


# 3.
LHS = (AB).T
RHS = B.T.dot(A.T)

print("\n3. (AB)^T:\n", LHS)
print("\n   B^T A^T:\n", RHS)
print("\nIs (AB)^T = B^T A^T ?", np.allclose(LHS, RHS))


# PART 2: SOLVE SYSTEM OF EQUATIONS
# System:
# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9

A2 = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3,  1, -1]
])

B2 = np.array([-1, -3, 9])

print("\n\nSystem of Equations in Matrix Form:")
print("A =\n", A2)
print("B =\n", B2)

# Solve using inverse method: X = A^{-1} B
A2_inv = np.linalg.inv(A2)
X_solution = A2_inv.dot(B2)

print("\nSolution using Inverse Method (X = A^{-1}B):")
print("x, y, z =", X_solution)


# ---------------------------------------
# PART 3: Solve using numpy.linalg.solve
# ---------------------------------------

X_solve = np.linalg.solve(A2, B2)

print("\nSolution using np.linalg.solve:")
print("x, y, z =", X_solve)

# Check if both methods match
print("\nDo both solutions match?", np.allclose(X_solution, X_solve))
```
